```python
import pandas as pd
```

```python
df = pd.read_csv('rainfall.csv')
df.head()
```

```python
df[' Statistics'] = df[' Statistics'].str.replace(' Average', '')
```

```python
def panda_strip(x):
    r =[]
    for y in x:
        if isinstance(y, str):
            y = y.strip()

        r.append(y)
    return pd.Series(r)

df = df.apply(lambda x: panda_strip(x))
```

```python
df.to_csv('rainfall.csv', index=False)
```
